```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

df= pd.read_csv(r"C:\Users\akash\OneDrive\Documents\AIML_projs\teams.csv")
teams=df[ ['team','country','year','athletes','age','prev_medals','medals'] ]

teams
```

```python
# to find rows with missing values :
teams[teams.isnull().any(axis=1)]

teams=teams.dropna()


train_set= teams[ teams['year']<2012].copy()
test_set = teams[ teams['year']>=2012].copy()
```

```python
train_set
```

```python
test_set
```

## Model Training

```python
reg=LinearRegression()
predictors=['athletes','prev_medals']
target=['medals']
reg.fit(train_set[predictors],train_set[target])
```

```python

predictions= reg.predict(test_set[predictors])
test_set['predictions']= predictions

test_set.loc[test_set['predictions']<0,'predictions']=0
test_set['predictions']=test_set['predictions'].round()
error = mean_absolute_error(test_set['medals'],test_set['predictions'])

```
